**Problem:**

Load a dataset and display the first 5 and last 5 rows. Use df.info() to get the DataFrame’s structure.

```python
# Importing Libraries
from datasets import load_dataset
import pandas as pd

# Loading Dataset
dataset = load_dataset("lukebarousse/data_jobs")
df = dataset['train'].to_pandas()

print(df.head())
print(df.tail())
print(df.info())
```

**Problem:**

Access the "salary_rate" column and calculate its mean. Also, find how many unique job titles are there.

```python
mean_salary = df.salary_year_avg.mean()
print(mean_salary)

unique_job_titles = df.job_title_short.unique()
print(unique_job_titles)
```

**Problem:**

Filter and display all rows where `salary_year_avg > 100000` and the job title contains `Data Scientist`.

```python
df[(df['job_title_short'] == 'Data Scientist') & (df['salary_year_avg'] > 100_000)]
```

**Problem:**

Remove all rows where the `job_skills` column is missing. Then, check how many rows are left in the DataFrame.

```python
cleaned_df = df[df['job_skills'].notna()]
print(len(cleaned_df))
```

**Problem**

Extract only those rows where the job is in `United States` and the work is remote `job_work_from_home == True`.

```python
df[(df['job_country'] == 'United States') & (df['job_work_from_home'] == True)]
```
