**Problem 3: Adding Labels and Title**

Question: Modify the previous line plot to include labels for the x-axis and y-axis, and add a title to the plot.

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [1, 2, 3, 4]

plt.plot(x, y)
plt.title("Simple Line Chart")
plt.xlabel("x-axis label")
plt.ylabel("y-axis label")
plt.show()
```

**Problem 4: Plotting Multiple Lines**

Question: Create a plot with two lines. The first line should use the data points x = [1, 2, 3, 4], y = [1, 4, 9, 16] and the second line should use the data points x = [1, 2, 3, 4], y = [2, 4, 6, 8]. Display the plot.

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y1 = [1, 4, 9, 16]
y2 = [2, 4, 6, 8]

plt.plot(x, y1, label= 'y = x^2')
plt.plot(x, y2, label= 'y = x*2')
plt.title("Multiple Line Chart")
plt.xlabel("x-label")
plt.ylabel("y-label")
plt.show()
```

**Problem 5: Creating a Bar Chart**

Question: Create a bar chart with the categories ['A', 'B', 'C', 'D'] and their corresponding values [10, 20, 15, 25]. Display the chart.

```python
import matplotlib.pyplot as plt

categories = ['A', 'B', 'C', 'D']
values = [10, 20, 15, 25]

plt.bar(categories, values)
plt.title("Simple Bar Chart")
plt.xlabel("Category")
plt.ylabel("Values")
plt.show()
```
