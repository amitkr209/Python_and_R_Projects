## Create a New Column

```python
# Create a new column named 'job_posted_month' with month_number in it.
df['job_posted_month'] = df['job_posted_date'].dt.month
```

```python
df.head()
```

```python
# creating a new column called 'Is Data Analyst' and saying if the column job_title_short is equal to 'Data Analyst' 
# then then return 1 if not, return 0. It does this by using astype(int).
df['Is Data Analyst'] = (df['job_title_short'] == 'Data Analyst').astype(int)

df.head()
```

```python
# Displaying only the data Analyst jobs
df[df['Is Data Analyst'] > 0]
```

```python
# Sort the data by job_posted_date in decending order
df.sort_values(by='job_posted_date', ascending=False)
```

```python
# Now, fix the decending order sort using inplace
# Method 1
df.sort_values(by='job_posted_date', ascending=False, inplace=True)
```

## Drop Data
Use `drop()` if you want to drop (delete) either a column or row in your database.

#### Drop Columns

Let's drop the column `salary_hour_avg`, this will have an axis of 1 (since we're dropping a column).

```python
df.drop(columns='salary_hour_avg')
```

Let's drop the column `job_posted_month`, this will have an axis of 1 and fix them.

```python
df.drop(['job_posted_month', 'Is Data Analyst'], axis=1, inplace=True)
```

```python
df.head()
```

#### Remove NA

Let's cleanup our `salary_year_avg` column by removing the `NaN` values in this column.

```python
df.dropna(subset=['salary_year_avg'], inplace= True)
```
