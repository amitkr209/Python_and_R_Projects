**Question:**

Graph the average salary `salary_year_avg` by `job_title_short` as a horizontal bar chart. Order it from the highest salary to the lowest.

Include a title, and labels for the x and y axis.

```python
# Loading Libraries
from datasets import load_dataset
import pandas as pd
import matplotlib.pyplot as plt

# Loading Data
dataset = load_dataset("lukebarousse/data_jobs")
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
job_salary = df.groupby('job_title_short')['salary_year_avg'].median().sort_values()

job_salary
```

```python
job_salary.plot(kind='barh')
plt.title("Median Salary by Job Title", fontsize=15)
plt.gca().xaxis.set_major_formatter(plt.FuncFormatter(lambda x, pos: f'${int(x/1000)}K'))
plt.xlabel("Median Salary ($USD)")
plt.ylabel("")
plt.show()
```
