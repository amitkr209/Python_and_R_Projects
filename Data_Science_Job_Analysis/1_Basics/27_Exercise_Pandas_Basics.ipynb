# AI Generated Questions

**Problem 1: Filtering DataFrame Rows**

Question: Filter the DataFrame to only include job postings from the `United Kingdom`.

```python
# Importing Libraries
from datasets import load_dataset
import pandas as pd

# Loading dataset
dataset = load_dataset("lukebarousse/data_jobs")
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])

# Filtering job postings from the United Kingdom
uk_df = df[df['job_country'] == 'United Kingdom']
uk_df
```

**Problem 2: Calculating Average Salary**

Question: Calculate the `average yearly` salary for `Data Engineer` job titles in the `United States`.

```python
# Filtering job postings from the United States
us_df = df[df['job_country'] == 'United States']

# Calculating the average yearly salary for Data Engineer job titles
average_salary_data_engineer = us_df[us_df['job_title_short'] == 'Data Engineer']['salary_year_avg'].mean()
average_salary_data_engineer
```

**Problem 3: Counting Job Postings**

Question: Count the number of `Data Scientist` job postings in the `United States`.

```python
data_scientist_count = us_df[us_df['job_title_short'] == 'Data Scientist'].shape[0]
data_scientist_count
```

**Problem 4: Determining Salary Range**

Question: Determine the `minimum` and `maximum` yearly salaries for `Business Analyst` job titles.

```python
salary_range_business_analyst = us_df[us_df['job_title_short'] == 'Business Analyst']['salary_year_avg'].agg(['min', 'max'])
salary_range_business_analyst
```

**Problem 5: Grouping and Aggregating Data**

Question: Group the DataFrame by `job_title_short` and find the `average`, `minimum`, and `maximum` yearly salaries for each job title in the `United States`.

```python
salary_stats_by_title = us_df.groupby('job_title_short')['salary_year_avg'].agg(['mean', 'min', 'max']).sort_values(by='mean', ascending=False)
salary_stats_by_title
```
