# Practise Exercise

```python
# Importing Libraries
from datasets import load_dataset
import pandas as pd
import matplotlib.pyplot as plt

# Loading Dataset
dataset = load_dataset("lukebarousse/data_jobs")
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
# Job Country Bar Chart
job_country = df.job_country.value_counts().head(5)

plt.bar(job_country.index, job_country.values)
plt.show()
```

```python
# Average Salary Horizontal Bar Chart
average_salary = df.groupby('job_title_short')['salary_year_avg'].mean()
average_salary = average_salary.sort_values(ascending=True)

plt.barh(average_salary.index, average_salary.values)
plt.show()
```

```python
# Average Salary Trend Line Plot
df['job_posted_month'] = df['job_posted_date'].dt.month
average_salary_by_month = df.groupby('job_posted_month')['salary_year_avg'].mean()

plt.plot(average_salary_by_month.index, average_salary_by_month.values)
plt.show()
```
