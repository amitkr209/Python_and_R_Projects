```python
# Importing Libraries
from datasets import load_dataset
import pandas as pd
import matplotlib.pyplot as plt

# Loading data
dataset = load_dataset("lukebarousse/data_jobs")
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

## Bar Charts - Counts of Job Postings

```python
job_counts = df.job_title_short.value_counts()

job_counts
```

```python
plt.bar(job_counts.index, job_counts.values)
plt.title("Postings of Job title")
plt.ylabel("Count of Job Postings")
plt.xticks(rotation=45, ha='right')
plt.show()
```

## Plot using Pandas

```python
job_counts.plot(kind='bar')
plt.title("Postings of Job title", fontsize=15)
plt.xlabel("")
plt.ylabel("Count of Job Postings")
plt.xticks(rotation=45, ha='right')
plt.show()
```

## Plot Method with Dataframe

```python
df[['job_posted_date', 'salary_year_avg']].dropna(subset=['salary_year_avg'])
```

```python
df.plot(x='job_posted_date', y='salary_year_avg', kind='line')
```

# Practise Problems

```python
# Job Posting Location Bar Plot
df.job_country.value_counts().head(10).plot(kind='bar')
plt.title("Job Posting Location", fontsize=15)
plt.xlabel("")
plt.ylabel("Counts of Job Postings")
plt.show()
```

```python
# Job Postings Trend Line Plot
df['job_posted_month'] = df.job_posted_date.dt.month

df['job_posted_month'].value_counts().sort_index(ascending=True).plot(kind='line')
plt.title("Job Postings Trend", fontsize=15)
plt.xlabel("")
plt.ylabel("Counts of Job Postings")
plt.show()
```

```python
# Top Job Locations
df.job_location.value_counts().head(10).plot(kind='bar')
plt.title("Top Job Locations", fontsize=15)
plt.xlabel("")
plt.ylabel("Counts of Job Postings")
plt.xticks(rotation=45, ha='right')
plt.show()
```
