```python
# Importing Libraries
from datasets import load_dataset
import pandas as pd
import matplotlib.pyplot as plt

# Loading Data
dataset = load_dataset("lukebarousse/data_jobs")
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

### Bar Charts - Count of Job Postings

```python
job_counts = df.job_title_short.value_counts()
job_counts
```

```python
plt.bar(job_counts.index, job_counts.values)
plt.title("Postings by Job Title", fontsize=15)
plt.ylabel("Count of Job Postings")
plt.xticks(rotation=45, ha='right')
plt.show()
```

### Practise Questions

#### Job Country Bar Chart with labels

```python
job_country = df['job_country'].value_counts().head(10)
job_country
```

```python
plt.bar(job_country.index, job_country.values)
plt.title('Posting by Job Country')
plt.ylabel('Counts of Job Postings')
plt.xlabel('Country')
plt.xticks(rotation=45, ha='right')
plt.show()
```

#### Average Salary Horizontal Bar Chart

```python
avg_salary = df.groupby('job_title_short')['salary_year_avg'].mean()
avg_salary = avg_salary.sort_values(ascending=True)
avg_salary
```

```python
plt.barh(avg_salary.index, avg_salary)
plt.title('Average Salary of Job Titles', fontsize=15)
plt.xlabel('Average Salary ($USD)')
plt.gca().xaxis.set_major_formatter(plt.FuncFormatter(lambda x, pos: f'{int(x/1000)}K'))
plt.show()
```

#### Average Salary Trend Line Plot

```python
df['job_posted_month'] = df['job_posted_date'].dt.month

month_avg = df.groupby('job_posted_month')['salary_year_avg'].mean()
month_avg
```

```python
plt.plot(month_avg.index, month_avg.values)
plt.title('Average Salary Trends by Month')
plt.ylabel('Average Salary ($USD)')
plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda y, pos: f'${int(y/1000)}K'))
plt.show()
```

**Problem 5: Customizing Tick Marks**

Question: Create a bar chart for the job counts with the following customizations:

- Title: "Job Postings by Job Title"
- Y-axis label: "Count of Postings"
- Set custom x-ticks to show only "Data Analyst", "Data Engineer", and "Data Scientist"
- Rotate the x-axis labels by 90 degrees

```python
# Importing Libraries
from datasets import load_dataset
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Loading Data
dataset = load_dataset("lukebarousse/data_jobs")
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])

job_counts = df['job_title_short'].value_counts()

plt.bar(job_counts.index, job_counts.values)
plt.title('Job Postings by Job Tilte', fontsize=15)
plt.ylabel('Count of Postings ->')
plt.xticks(ticks=[0, 1, 2], labels=['Data Analyst', 'Data Engineer', 'Data Scientist'], rotation=90, ha='center', va='top')
plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda y, pos: f'{int(y/1000)}K'))
sns.despine()
plt.show()
```
