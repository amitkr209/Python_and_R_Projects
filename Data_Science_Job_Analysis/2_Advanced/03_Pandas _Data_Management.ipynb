```python
# Importing Library
from datasets import load_dataset
import numpy as np
import pandas as pd

# Loading Datasets
dataset = load_dataset("lukebarousse/data_jobs")
df = dataset['train'].to_pandas()
```

```python
df.head()
```

```python
# Data Cleaning
df['job_posted_date'] = pd.to_datetime(df.job_posted_date)

# Creating the copy of the original df
df_original = df.copy()
```

```python
df_altered = df_original.copy()
```

```python
# Calculate the median salary
median_salary = df_altered['salary_year_avg'].median()

# Filling the missing values into the df_altered.
df_altered['salary_year_avg'] = df_altered.salary_year_avg.fillna(median_salary)

# Inspect the altered dataframe
df_altered.loc[:10, 'salary_year_avg']
```

```python
df_altered.sample(5)
```
