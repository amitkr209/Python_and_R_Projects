```python
# Loading Library
from datasets import load_dataset
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Loading dataset
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Cleaning dataset
df['job_posted_date'] = pd.to_datetime(df.job_posted_date)
```

```python
df.head(3)
```

```python
# Get the current index of the DataFrame and its data type.
print(df.index)
print(df.index.dtype)
```

```python
# Set the name of the index to 'job_index'
# Verify that the index name has been set correctly.
df.index.name = 'job_index'
print(df.index.name)
```

```python
# Create a pivot table to get the median yearly salaries for different job titles.
pivot_table = df.pivot_table(values='salary_year_avg', index='job_title_short', aggfunc='median')

# Get the index name and data type of the pivot table.
print(pivot_table.index.name)
print(pivot_table.index.dtype)
```

```python
# Reset the index of the pivot table
pivot_table_reset = pivot_table.reset_index()

# Verify the changes
pivot_table_reset.head()
print(pivot_table.index.name)
```

```python
# Filter the DataFrame to get job postings in the 'United States'
df_us = df[df.job_country == 'United States']

# Inspect the first five rows of the filtered DataFrame.
df_us.head()

# Reset the index of the filtered DataFrame and 
df_us_reset = df_us.reset_index(drop=True)

# verify the changes
df_us_reset.head()
```

```python
# Set 'job_index' as the index for the filtered DataFrame
df_us_reset.index.name = 'job_index'

# Inspect the first five rows
df_us_reset.head()
```
