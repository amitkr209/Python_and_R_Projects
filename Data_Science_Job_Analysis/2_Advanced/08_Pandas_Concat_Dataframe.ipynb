```python
# Importing Library
from datasets import load_dataset
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Loading dataset
dataset = load_dataset("lukebarousse/data_jobs")
df = dataset['train'].to_pandas()

# Data Cleaning
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
# Sample dataset of job postings in January
job_postings_jan = pd.DataFrame({
    'job_id': [1, 2, 3, 4, 5],
    'job_title': ['Data Scientist', 'Data Analyst', 'Machine Learning Engineer', 'Data Scientist', 'Data Engineer'],
    'company': ['Company A', 'Company B', 'Company C', 'Company D', 'Company E'],
    'job_posted_date': pd.to_datetime(['2024-01-02', '2024-01-07', '2024-01-14', '2024-01-19', '2024-01-24'])
})

job_postings_jan
```

```python
# Sample dataset of job postings in February
job_postings_feb = pd.DataFrame({
    'job_id': [6, 7, 8, 9, 10],
    'job_title': ['Data Scientist', 'Data Analyst', 'Machine Learning Engineer', 'Data Scientist', 'Data Engineer'],
    'company': ['Company F', 'Company G', 'Company H', 'Company I', 'Company J'],
    'job_posted_date': pd.to_datetime(['2024-02-05', '2024-02-09', '2024-02-12', '2024-02-18', '2024-02-22'])
})

job_postings_feb
```

```python
# Merging the two dataframes
job_postings_jan.merge(job_postings_feb, on='job_title', suffixes=('_jan', '_feb'))
```

```python
# Concat the two dataframes
job_postings_combine = pd.concat([job_postings_jan, job_postings_feb])

job_postings_combine
```

```python
# While Concat the index, we can ignore the index
job_postings_combine = pd.concat([job_postings_jan, job_postings_feb], ignore_index=True)

job_postings_combine
```

```python
df['job_posted_month'] = df['job_posted_date'].dt.strftime('%b')

df.head()
```

```python
months = df['job_posted_month'].unique()

dict_months = {month : df[df['job_posted_month'] == month] for month in months}

dict_months['Jan']
```

```python
# You have quarterly sales data for different products. Create a MultiIndex DataFrame from these dataframes.

# Quarterly sales data
sales_q1 = pd.DataFrame({
    'product_id': [1, 2, 3],
    'sales_q1': [100, 200, 300]
})

sales_q2 = pd.DataFrame({
    'product_id': [1, 2, 3],
    'sales_q2': [110, 210, 310]
})

sales_q3 = pd.DataFrame({
    'product_id': [1, 2, 3],
    'sales_q3': [120, 220, 320]
})

sales_q4 = pd.DataFrame({
    'product_id': [1, 2, 3],
    'sales_q4': [130, 230, 330]
})
```

```python
sales_combined = pd.concat([sales_q1.set_index('product_id'),
                           sales_q2.set_index('product_id'),
                           sales_q3.set_index('product_id'),
                           sales_q4.set_index('product_id')],
                           axis = 1)
                           

print(sales_combined)
```
