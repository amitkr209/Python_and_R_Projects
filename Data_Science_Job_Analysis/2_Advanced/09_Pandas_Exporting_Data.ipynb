```python
# Importing Library
from datasets import load_dataset
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Loading data
dataset = load_dataset("lukebarousse/data_jobs")
df = dataset['train'].to_pandas()

# Data Cleaning
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df_DA = df[df['job_title_short'] == 'Data Analyst'].copy()

df_DA
```

```python
df_DA['job_posted_month'] = df['job_posted_date'].dt.strftime('%b')

df_DA.head()
```

```python
months = df_DA['job_posted_month'].unique()

months
```

```python
dict_months = {month : df_DA[df_DA['job_posted_month'] == month] for month in months}

dict_months['Jan']
```

```python
quater_1 = [dict_months['Jan'], dict_months['Feb'], dict_months['Mar']]

df_q1 = pd.concat(quater_1, ignore_index=True)

df_q1
```
