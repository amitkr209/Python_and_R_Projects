```python
# Importing Libraries
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
# Health Insurance Pie Chart

df['job_health_insurance'].value_counts().plot(kind='pie', startangle=90, autopct='%1i%%')
plt.title("Health Insurance Status")
plt.ylabel('')
plt.show()
```

```python
# Job Schedule Distribution

job_schedule = ['Full-time', 'Contractor', 'Internship']
df_plot = df[df['job_schedule_type'].isin(job_schedule)]['job_schedule_type'].value_counts()

df_plot.plot(
    kind='pie', 
    startangle=90, 
    autopct='%1.1f%%',
    explode=(0.1, 0)
    )
plt.title("Job Schedule Distribution", fontsize=15)
plt.ylabel('')
plt.show()
```

```python
fig, ax = plt.subplots()

ax.pie(df['job_no_degree_mention'].value_counts(), startangle=90, autopct='%1.1f%%', labels=['False', 'True'])
plt.title("Degree Mentioned Pie Chart", fontsize=15)
plt.show()
```

```python
df_DA = df[df['job_title_short'] == 'Data Analyst'].copy()
df_DA.head()
```

```python
# Count True and False values for degree requirement
job_no_degree = df['job_no_degree_mention'].value_counts()

# Plotting the Pie Chart
plt.figure(figsize=(6, 6))
plt.pie(job_no_degree,
         startangle=90, 
         autopct='%1.1f%%',
         explode=(0.2, 0), 
         labels=['False', 'True'], 
         colors=['lightblue', 'orange']
         )
plt.title("Degree Requirement for Data Analyst Jobs", fontsize=15)
plt.show()
```

```python
fig, ax = plt.subplots(1, 3, figsize=(12, 6))

dict_column = {
    'job_work_from_home' : 'Work from Home Offered',
    'job_no_degree_mention' : 'Degree Requirement',
    'job_health_insurance' : 'Health Insurance Offered'
}

for i, (column, title) in enumerate(dict_column.items()):
    ax[i].pie(df[column].value_counts(), 
              startangle=90, autopct='%1.1f%%', 
              labels={'False', 'True'},
              explode=(0.1, 0),
              colors=['lightblue', 'orange']
              )
    ax[i].set_title(title, fontsize=15)
```
